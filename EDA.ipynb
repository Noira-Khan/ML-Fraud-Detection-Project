```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def run_eda(data_path="CreditCard.csv", output_dir="eda_outputs"):
   
    
    os.makedirs(output_dir, exist_ok=True)
    log_file_path = os.path.join(output_dir, "eda_summary_metrics.txt")
    
    print(f"Initializing EDA Pipeline. Loading dataset from: {data_path}")
    try:
        df = pd.read_csv(data_path)
    except FileNotFoundError:
        print(f"[!] Error: Dataset not found at {data_path}. Verify your directory structure.")
        return

    with open(log_file_path, "w") as log:
        log.write("      CREDIT CARD FRAUD DETECTION - EDA     \n")

        log.write("--- 1. DATA SHAPE & CLEANLINESS---\n")
        log.write(f"Total Observations (Rows): {df.shape[0]}\n")
        log.write(f"Total Features (Columns): {df.shape[1]}\n")
        missing_values = df.isnull().sum().sum()
        log.write(f"Missing Values Count: {missing_values}\n\n")

    
        log.write("--- 2. CLASS DISTRIBUTION ANALYSIS ---\n")
        class_counts = df['Class'].value_counts()
        class_pct = df['Class'].value_counts(normalize=True) * 100
        log.write(f"Legitimate Transactions (Class 0): {class_counts[0]} ({class_pct[0]:.4f}%)\n")
        log.write(f"Fraudulent Transactions (Class 1): {class_counts[1]} ({class_pct[1]:.4f}%)\n\n")

        log.write("--- 3. DESCRIPTIVE STATISTICS BY CLASS ---\n")
        stats_by_class = df.groupby('Class')['Amount'].describe()
        log.write(f"{stats_by_class.to_string()}\n\n")

        log.write("--- 4. TOP LINEAR CORRELATIONS WITH TARGET (CLASS) ---\n")
        correlations = df.corr()['Class'].sort_values()
        log.write("Strongest Negative Correlations with Fraud:\n")
        log.write(f"{correlations.head(5).to_string()}\n\n")
        log.write("Strongest Positive Correlations with Fraud:\n")
        log.write(f"{correlations.tail(6).to_string()}\n\n")

    print(f"Text metrics calculated successfully. Log saved to: {log_file_path}")

        ## generating visualizations
    print("Generating statistical distributions and visualizations...")
    sns.set_theme(style="whitegrid")
    fig = plt.figure(figsize=(20, 12))
    grid = plt.GridSpec(2, 2, wspace=0.3, hspace=0.3)

    # Plot A: Boxplot of Transaction Amount with Logarithmic Scaling
    ax0 = fig.add_subplot(grid[0, 0])
    sns.boxplot(x='Class', y='Amount', data=df, ax=ax0, palette="coolwarm", hue='Class', legend=False)
    ax0.set_yscale('log')  # Resolves heavily skewed feature variance
    ax0.set_title('Distribution of Transaction Amounts (Log Scale)', fontsize=14, fontweight='bold')
    ax0.set_xticklabels(['Legit (0)', 'Fraud (1)'])
    ax0.set_ylabel('Amount ($ Log Scale)')

    # Plot B: Kernel Density Estimate (KDE) for Transaction Time
    ax1 = fig.add_subplot(grid[0, 1])
    sns.kdeplot(df[df['Class'] == 0]['Time'], label='Legit (0)', fill=True, color='teal', ax=ax1, alpha=0.4)
    sns.kdeplot(df[df['Class'] == 1]['Time'], label='Fraud (1)', fill=True, color='crimson', ax=ax1, alpha=0.4)
    ax1.set_title('Transaction Density over Time Elapsed', fontsize=14, fontweight='bold')
    ax1.set_xlabel('Seconds Elapsed Since First Transaction')
    ax1.set_ylabel('Density')
    ax1.set_xlim(df['Time'].min(), df['Time'].max())
    ax1.legend()

    # Plot C: Comprehensive Feature Correlation Heatmap Matrix
    ax2 = fig.add_subplot(grid[1, :])
    correlation_matrix = df.corr()
    sns.heatmap(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1, ax=ax2, 
                cbar_kws={'label': 'Pearson Correlation Coefficient'}, xticklabels=True, yticklabels=True)
    ax2.set_title('Global Feature Correlation Matrix (Linear Dependency)', fontsize=14, fontweight='bold')
    ax2.tick_params(axis='both', which='major', labelsize=9)

    # Save and compress the graphs
    plot_output_path = os.path.join(output_dir, "eda_distribution_plots.png")
    plt.savefig(plot_output_path, dpi=300, bbox_inches='tight')
    plt.close()
    print(f"Visual plots successfully exported to: {plot_output_path}")
    print("EDA Pipeline completed successfully.")

if __name__ == "__main__":
    run_eda()
```
