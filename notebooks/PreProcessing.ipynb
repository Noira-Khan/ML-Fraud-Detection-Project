```python
import pandas as pd
from sklearn.preprocessing import RobustScaler

def preprocess_data(file_path):
    print("Loading credit card dataset...")
    df = pd.read_csv(file_path)
    print(f"Original dataset shape: {df.shape}")

   
    # It uses Median and IQR, making it immune to extreme transaction amount outliers
    scaler = RobustScaler()

    print("Scaling 'Amount' and 'Time' columns")
    
    # Double brackets [['column']] are used because scikit-learn scalers require 2D arrays
    df['scaled_amount'] = scaler.fit_transform(df[['Amount']])
    df['scaled_time'] = scaler.fit_transform(df[['Time']])

    # 4. Drop the original, unscaled 'Time' and 'Amount' columns
    df.drop(['Time', 'Amount'], axis=1, inplace=True)

    scaled_amount = df.pop('scaled_amount')
    scaled_time = df.pop('scaled_time')
    
    df.insert(0, 'scaled_amount', scaled_amount)
    df.insert(1, 'scaled_time', scaled_time)

    print("Preprocessing complete!")
    return df

if __name__ == "__main__":
    csv_path = 'CreditCard.csv' 
    
    processed_df = preprocess_data(csv_path)

    print("\n--- Summary Statistics for Scaled Columns ---")
    print(processed_df[['scaled_amount', 'scaled_time']].describe())
    
    print("\n--- First 5 Rows of Processed Dataset ---")
    print(processed_df.head())

   
```
